```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
import seaborn as sns
```

```python
def get_excel_columns(n_cols):
    col_list = []
    n = n_cols
    alpha_list = [chr(c) for c in range(ord('A'), ord('Z') + 1)]
    alpha_list.insert(0,'dummy')
    for i in range(1,n+1):
        col_i = ""
        index_list=[]
        p=[]
        while i>26:
            k = i%26
            i = i//26
            if k==0:
                index_list.append(26)
                i = i-1
            else:
                index_list.append(k)
        index_list.append(int(i))
        index_list.reverse()
        for i in index_list:
            p.append(alpha_list[i])
            col_i = "".join(p)
        col_list.append(col_i)
    return col_list
```

```python
df = pd.read_csv("winconsin_breast_data.csv", header = None)
df.columns = get_excel_columns(len(df.columns))
df
```

1. train_test_split(test_size = 0.3) (Extract only the inliers from X_train)

```python
# data frame of all INLIERS
df_inliers = df.loc[df.iloc[:,-1] == 'n']
df_inliers
```

```python
y = df.iloc[:,-1]
y = LabelEncoder().fit_transform(y)
y
```

```python
X = df.iloc[:, 0:-1]
X = StandardScaler().fit(X).transform(X.astype(float))
X[0:5]
```

```python
X_inliers = df_inliers.iloc[:, 0:-1]
X_inliers = StandardScaler().fit(X_inliers).transform(X_inliers.astype(float))
X_inliers
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
print("Train Data: ", X_train.shape, y_train.shape)
print("Test Data: ", X_test.shape, y_test.shape)
```

# Case 1

```python
accuracy1 = []
plt.figure(figsize = (10,6))
high = 100
for i in range(1,high+1):
    clf = LocalOutlierFactor(n_neighbors = i)
    y_pred = clf.fit_predict(X_train)
    y_pred_enc = np.where(y_pred == -1, 1,0)
    accuracy1.append(metrics.accuracy_score(y_train, y_pred_enc)*100)
sns.lineplot(range(1, high + 1), accuracy1)
```

# Case 2

```python
accuracy2 = []
plt.figure(figsize = (10,6))
high = 100
for i in range(1,high+1):
    clf = LocalOutlierFactor(n_neighbors = i)
    y_pred = clf.fit_predict(X)
    y_pred_enc = np.where(y_pred == -1, 1,0)
    accuracy2.append(metrics.accuracy_score(y, y_pred_enc)*100)
sns.lineplot(range(1, high + 1), accuracy2)
```

# Case 3a

```python
accuracy3a = []
plt.figure(figsize = (10,6))
high = 100
for i in range(1,high+1):
    clf = LocalOutlierFactor(n_neighbors = i, novelty = True, contamination = 'auto')
    clf.fit(X_inliers)
    y_pred = clf.predict(X)
    y_pred_enc = np.where(y_pred == -1, 1,0)
    accuracy3a.append(metrics.accuracy_score(y, y_pred_enc)*100)
sns.lineplot(range(1, high + 1), accuracy3a)
```

# Case 3b

```python
accuracy3b = []
plt.figure(figsize = (10,6))
high = 100
low = 1
for i in range(low,high+1):
    clf = LocalOutlierFactor(n_neighbors = i, novelty = True, contamination = 0.1)
    clf.fit(X_inliers)
    y_pred = clf.predict(X)
    y_pred_enc = np.where(y_pred == -1, 1,0)
    accuracy3b.append(metrics.accuracy_score(y, y_pred_enc)*100)
sns.lineplot(range(low, high + 1), accuracy3b)
```

# Case 4a

```python
# extract only the inliers from the training set (X_train)

y_train_list = []
for i in y_train:
    y_train_list.append(i)
j=0
X_train_list = []
for i in X_train:
    m = []
    m = i.tolist()
    m.append(y_train_list[j])
    X_train_list.append(m)
    j = j+1
df_rejoined = pd.DataFrame(X_train_list)
#df_rejoined[30].value_counts()
df_rejoined_inliers = df_rejoined.loc[df_rejoined.iloc[:,-1] == 0]
X_train_inliers = df_rejoined_inliers.iloc[:, 0:-1].to_numpy()
X_train_inliers
```

```python
accuracy4a = []
plt.figure(figsize = (10,6))
high = 100
low = 1
for i in range(low,high+1):
    clf = LocalOutlierFactor(n_neighbors = i, novelty = True, contamination = 'auto')
    clf.fit(X_train_inliers)
    y_pred = clf.predict(X_test)
    y_pred_enc = np.where(y_pred == -1, 1,0)
    accuracy4a.append(metrics.accuracy_score(y_test, y_pred_enc)*100)
sns.lineplot(range(low, high + 1), accuracy4a)
```
