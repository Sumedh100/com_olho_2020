# KNN on http requests data

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

# Load Data

```python
df = pd.read_csv("HBOS_requests.csv", header = None)
print(df.shape)
df.head()
```

```python
def get_excel_columns(n_cols):
    col_list = []
    n = n_cols
    alpha_list = [chr(c) for c in range(ord('A'), ord('Z') + 1)]
    alpha_list.insert(0,'dummy')
    for i in range(1,n+1):
        col_i = ""
        index_list=[]
        p=[]
        while i>26:
            k = i%26
            i = i//26
            if k==0:
                index_list.append(26)
                i = i-1
            else:
                index_list.append(k)
        index_list.append(int(i))
        index_list.reverse()
        for i in index_list:
            p.append(alpha_list[i])
            col_i = "".join(p)
        col_list.append(col_i)
    return col_list
```

```python
n_cols = len(df.columns)
n_cols
```

```python
df.columns = get_excel_columns(n_cols)
df.head()
```

## Train - Test split

```python
X = df[get_excel_columns(len(df.columns)-1)].values.astype(float)
X[0:5]
```

```python
y = df.iloc[:,-1]
print(len(y))
y[0:5]
```

```python
X = preprocessing.StandardScaler().fit(X).transform(X.astype(float))
X[0:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 4)
print("Train Data: ", X_train.shape, y_train.shape)
print("Test Data: ", X_test.shape, y_test.shape)
```

### Classification using KNN 

```python
error_rate = []
for i in range(1,5):
    KNN = KNeighborsClassifier(n_neighbors = i)
    KNN.fit(X_train, y_train)
    y_hat = KNN.predict(X_test)
    error_rate.append(np.mean(y_hat != y_test))
error_rate
```

```python
rofl = sns.lineplot(range(1,40), error_rate, markers = True, dashes = False)
```
